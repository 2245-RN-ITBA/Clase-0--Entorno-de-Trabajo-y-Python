````python
class ClaseMath:
    def __init__(self, a, b=5):
        self.a = a
        self.b = b
        print("Clase inicializada")
    def suma(self):
        return self.a + self.b
    def resta(self):
        return self.a - self.b
    def __multip(self):
        return a*b
    def mult(self):
        return self.__multip()
    
````
